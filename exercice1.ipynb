## Les éléments de base
- Recherche de mot ou de lettre dans un text peut import le volume du texte

```python
# Recherhe dans le text, vous pouvez mettre "r" comme ne pas le mettre dans le print
print(re.findall(r't', text))  # ce ci veut qu'on cherche tous les <t> dans le text
print(re.findall('l', text))   # ce ci veut qu'on cherche tous les <l> dans le text
print(re.findall('n', text))   # ce ci veut qu'on cherche tous les <n> dans le text
print(re.findall('le', text))   # ce ci veut qu'on cherche tous les <le> dans le text
```

# LES DISJONCTION
1. Caractères dans les crochets [ ]
* Les crochets sont utilisés pour disigner une liste de Caractères possible à une position données dans le motif de recherche *

```python
# Un exemple d'application
motif = "Attieke, attiéke, Banane, banane" 
# description du motif, le debut des mots sont en miniscule et majuscule, nous allons faire l'affichage avec [ ]

re.findall(r'Attieke', motif) # ici on specifie le type qu'on veut extraire ici: 'Attieke'

# Pour recuperer les initiales Majuscule et Minuscule on utilise les crochets comme suit :

print("Attieke en majuscule et minuscule : ", re.findall('[Aa]tti[eé]ke', motif)) # identifié la position de la lettre
print("Banane en majuscule et minuscule : ", re.findall('[Bb]anane', motif))
```

## Négation de disjonction
- ^ signifie la négation lorsqu'il est le premier caractère entre les crochets

```python
# Cas d'usage de la négation appliquons le sur les mots à 3 lettres
neg = '123 adb2 ABB55B???'

txt = 'bo boo boooo boooooooooo booooooooooooooo booooooooooooooooooo'
# Affichage
print("Tous les caractères  sauf le ? : ", re.findall(r'[^?]', neg), "\n") # Tous les caractères  sauf le ?, recupère aussi les caractères vides.
print("Au moins 1 fois la dernière lettre : ", re.findall(r'bo+', txt), "\n") # Au moins une fois la dernière lettre, il recupère que les bo.
print("Au moins 0 (>=) fois la dernière lettre : ", re.findall(r'bo*', txt), "\n") # Au moins une fois la dernière lettre, il recupère que les bo.
```

## Les alias pratiques
* \d   ==== [0-9]   Tous les chiffres de 0 à 9
* \D   ==== [^0-9]      Toutes les caractères non numériques
* \w   ==== [a-zA-Z0-9_]     Tous les caractères alphanumériques ou _
* \W   ==== [^\w]   Tous les caractères non alphanumériques ou sans _
* \s  ====  [\t\n\r\f]  Tout espace (retour a la ligne tabulation)
* \S   ==== [^\s]   Toutes les caractères autres que les spaces

```python
# Simplifions l'utilisation des numéros de téléphones

num_orange = r'07[\d]{8}'
print("Les numeros oranges sont :", re.findall(num_orange, plag))
```

```python
text = \
"""
Ami, frère, collègue, jeune, âne,
Attiéké Poisson Fumé
attiéké poisson fumé
0534236798
0734236790
0704236709
0504246780
07042467
"""
# Application de la logique des alias pratiques |
car = r'07\d{8}|\w{4}'
mot = r'[Aa]ttiéké|[Pp]oisson|[Ff]umé'
print("Les numeros oranges et toutes caractères de 4 lettres sont :", re.findall(car, text),"\n")
print("Les mots attieke, poisson et fumé sont :", re.findall(mot, text))
```

## Début et fin de ligne avec (^)-($)

* ^ désigne le début de ligne: 
* ^[A-Z] --> toute majuscule en début de ligne
* ^[^a-zA-Z]: Tout caractère non alphabétique en début de ligne
* .$: Tout caractère en fin de ligne
* \.$: Tout point en fin de ligne

```python
# Cas pratiques
# Début et fin de ligne
motif_ = \
""" 
En Python, les expressions régulières (regex) sont gérées par le module `re`,
qui permet de trouver, d'extraire ou de remplacer des motifs de texte.
Elles sont utiles pour des tâches comme la validation de mots de passe
ou la recherche de motifs spécifiques dans des chaînes.
Mais aussi, elles servent à faire du scraping de texte.
Pour les utiliser, il faut d'abord importer le module `re`,
puis employer des fonctions comme `re.search()` pour trouver une occurrence,
`re.findall()` pour trouver toutes les occurrences,
ou `re.sub()` pour remplacer les motifs.
"""

ligne = r'^[A-Z]' 
print("Tous les mots commençant par une majuscule : ", re.findall(ligne, motif_, re.MULTILINE))
fin = r'\.$'
print("Tous les mots finissant par un point : ", re.findall(fin, motif_, re.MULTILINE))
```

## Sélection où capture de groupe

# EXTRACTION D'EMAIL

- username: lettres, chiffres, .-_
- domaine: lettres, chiffres, -
- extension: 2-6 lettres

```python
# Le format général d'un adresse gmail username@domain.extension

emails_test = [
    "contact@example.ci",      # Valide
    "user.name@startup.co.ci", # Valide
    "user-name@startup-co.ci", # Valide
    'ik34@uvci.edu.ci',
    "invalid@",                # Invalide
    "@invalid.com",            # Invalide
    "no-at-sign.com",          # Invalide
    "user@domain",             # Invalide (pas d'extension)
    "user name@domain.com"     # Invalide (espace)
]
```

```python
# EXTRACTION D'EMAIL
regex_email = r'^[\w.-]+@[a-zA-Z0-9-.]+\.[a-z]{2,6}$' # Le validateur
for email in emails_test:
    print(re.findall(regex_email, email))
```

```python
# 
for email in emails_test:
    if re.match(regex_email, email):
        print(f"{email} \t est valide ! ")
    else:
        print(f"{email} est invalide")
```

# Format valide de numéros en côte d'ivoire
Les formats valide sont :
- +225 XX XX XX XX XX
- +225 XXXX  XXXXX X
- 00225 XXXXXXXXXX
- (00225) XXXXXXXXXX
- 225 XXXXXXXXXX

```python
# Les numeros commencent par 01,05,07
numeros_test = [
    "+225 07 08 09 10 11",     # Valide
    "+22507 08 09 10 11",      # Valide
    "(225)01 08 09 10 11",      # Valide
    "(00225)07 08 09 10 11",      # Valide
    "00225 05 08 09 10 11",      # Valide
    "+22507 08 09 10 11",      # Valide|
    "07 08 09 10 11",          # Valide
    "07080    9101    1",              # Valide
    "01080 9101 1",              # Valide
    "05 08 0910 11",              # Valide
    "25 30 40   50",             # Invalide (trop court)
    "(+225) 7 8 9 10 11",        # InValide
    "1234567890"               # Invalide (ne commence pas par 0 ou 2)
]
```

```python
# extraction
indix_num = r'\(?(?:\+|0{2})?225\)?'
num_ci = r'[02][157](?:\s*\d\s*){8}'
phone_num = rf'(?:{indix_num})?\s*{num_ci}'
for numero in numeros_test:
    print(re.findall(phone_num, numero))
```

```python
# autres boucles
for numero in numeros_test:
    if re.match(phone_num, numero):
        print(numero, ':', 'Ce numero est valide')
    else:
        print(numero, ':', 'Ce numero est invalide')
```
